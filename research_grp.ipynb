```python
import json
with open("train.json") as fin:
    train_data = json.load(fin)
    
with open("test.json") as fin:
    test_data = json.load(fin)
```

```python
# the number of different author_grp
grps = set()

for paper_id, paper_val in train_data.items():
    # the list author sorted
    authors_str_list = [str(author_id) for author_id in sorted(paper_val['author'])]
    grp_member_str = '_'.join(authors_str_list)
    grps.add(grp_member_str)
    
print(len(grps))
```

```python
res_grps = grps

for grp in res_grps:
    if not grp.endswith('_dup'):
        author_set = set(grp.split('_'))
        for other_grp in grps:
            other_author_set = set(other_grp.split('_'))
            if len(author_set) < len(other_author_set):
                if author_set.issubset(other_author_set):
                    res_grps.discard(grp)
                    res_grps.add(grp + '_dup')
                
print(len(res_grps))
```

```python
count = 0


for grp in res_grps:
    if grp.endswith('_dup'):
        count += 1
        print(grp)

print(count)
```

```python
final_grps = set()
for grp in res_grps:
    if not grp.endswith('_dup'):
        final_grps.add(grp)
        
print(len(final_grps))
```

```python
final_grps_lists = list()
for grp in final_grps:
    authors = grp.split('_')
    final_grps_lists.append(authors)
    
print(final_grps_lists)
```

```python
def make_prediction(target, coauthors, grps):
    for authors in grps:
        if str(target) in authors:
            for coauthor in coauthors:
                if str(coauthor) in authors:
                    return 1
    return 0
```

```python
predictions = []

for test_id, test_val in test_data.items():
    coauthors = test_val['coauthor']
    target = test_val['target']
    
    prediction = make_prediction(target, coauthors, final_grps_lists)
    if prediction >= 0 and prediction <= 1:
        predictions.append([test_id, prediction])
        

```

```python
print(len(predictions))
```

```python
import pandas as pd
predictions_df = pd.DataFrame(columns=['Id','Predicted'], data=predictions)
print(predictions_df)
print(sum(predictions_df['Predicted']))
```

```python
predictions_df.to_csv('predictions__.csv', index=False)
```
